**4. Explain the difference between stemming and lemmatization.**

Ans:
* Stemming:

- It cuts off prefixes or suffixes from words using simple rules.
- It does not consider the actual meaning of the word.
- The output may not be a real word.

Examples:

running → run
studies → studi
playing → play
Fast but less accurate.

* Lemmatization:

- It reduces words to their dictionary base form.
- It considers context and meaning using vocabulary and grammar rules.
- The output is always a valid word.

Examples:

running → run
better → good
studies → study
Slower but more accurate.

**Task 2:Build Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Remove emojis & special characters
    text = re.sub(r"[^\w\s]", "", text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words except 'no', 'not'
    filtered_tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    clean_sentence = " ".join(filtered_tokens)

    return filtered_tokens, clean_sentence
```

**Task 3: Stress Testing**

Test your function on at least 10 diverse sentences.


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Clean:", clean)
    print("-"*50)

    results.append(tokens)
```

**Task 4: Token Analytics**

For each processed sentence, compute:

Total number of tokens

Number of unique tokens

Average token length


```python
import numpy as np

for i, tokens in enumerate(results):
    if len(tokens) == 0:
        continue

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens])

    print(f"Sentence {i+1}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
    print("-"*40)
```

Analysis Answers:

Most noisy sentence: "Win $$$ now!!! Limited offer!!!"

Most meaningful tokens retained: "I am not happy with this"

**Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = [token for sublist in results for token in sublist]

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[:-6:-1])
```

**Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Example
output = full_pipeline(test_sentences)
print(output)
```
